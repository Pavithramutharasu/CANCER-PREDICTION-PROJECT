# CANCER PREDICTION

**Objective:-**
         Develop a machine learning model to predict the presence or risk of cancer in patients based on medical data. The goal is to create a reliable and accurate predictive tool that can aid in early detection and intervention, ultimately improving patient outcomes.

**Data Source:-**
           dataset="/content/Cancer_Data.csv"

```python
# STEP 1 : Import libraries
import numpy as np
import pandas as pd
```

```python
# STEP 2 : Import dataset
dataset = pd.read_csv("/content/Cancer_Data.csv")
```

```python
# Data Visualization
import seaborn as sns
sns.regplot(x= 'texture_mean',y='compactness_worst',data=dataset, logistic=True)
```

```python
import matplotlib.pyplot as plt
f,ax=plt.subplots(figsize = (18,18))
sns.heatmap( dataset.drop(['id', 'diagnosis','Unnamed: 32'],axis = 1).corr(),annot= True,linewidths=0.5,fmt = ".1f",ax=ax)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title('Heat Map (showing the correlation matrix)')
plt.savefig('graph.png')
plt.show()
```

```python
dataset.columns
```

```python
# Define Target Variable (y) and Feature Variables (X)
y = dataset["diagnosis"]
X = dataset.drop(['id', 'diagnosis','Unnamed: 32'],axis = 1)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.7,random_state = 2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Modeling
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
```

```python
# Fit Model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Predict Model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Model Evaluation
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print("CONFUSION MATRIX\n")
confusion_matrix(y_test,y_pred)
```

```python
print("ACCURACY SCORE\n")
accuracy_score(y_test,y_pred)
```

```python
print("CLASSIFICATION REPORT\n")
print(classification_report(y_test,y_pred))
```
